```python
import numpy as np
import pandas as pd

```

```python
data = pd.read_csv('data.csv', header=[0,1], index_col=0)
data.head()
```

```python
syms = list(set([x[0] for x in data.columns]))
syms.sort()
syms[-1]
```

```python
change_dict={}
for sym in syms:
    dates = {}
    for row in data[sym].iterrows():
        dates[row[0]] = row[1]['4. close'] - row[1]['1. open']
    change_dict[sym] = dates
 
```

```python
change_df = pd.DataFrame(change_dict)
change_df.head()
```

```python
change_df.isna().sum()
```

```python
change_df.tail(20)
```

```python
change_df.to_csv('change.csv')
```

```python
percent_change_dict={}
for sym in syms:
    dates = {}
    for row in data[sym].iterrows():
        dates[row[0]] = 100 * (row[1]['4. close'] - row[1]['1. open'])/row[1]['1. open']
    percent_change_dict[sym] = dates
 
```

```python
percent_change_df = pd.DataFrame(percent_change_dict)
percent_change_df.head()
```

```python
percent_change_df.to_csv('percent_change.csv')
```

```python
range_dict={}
for sym in syms:
    dates = {}
    for row in data[sym].iterrows():
        dates[row[0]] = row[1]['2. high'] - row[1]['3. low']
    range_dict[sym] = dates
range_df = pd.DataFrame(range_dict)
range_df.to_csv('range.csv')
range_df.head()
```

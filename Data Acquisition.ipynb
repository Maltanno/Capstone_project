```python
# list from: https://www.nasdaq.com/market-activity/stocks/screener
idf = pd.read_csv('nasdaq_screener_1616319945743.csv')
idf.head()
```

```python
syms = idf.Symbol
```

```python
data = []
errors = []
```

```python
for sym in syms[:500]:
    try:
        d = ts.get_daily_adjusted(sym, outputsize='full')
        data.append(d)
    except:
        errors.append(sym)
    time.sleep(20)
```

```python
ddf = data[0][0].copy()
ddf.columns = [[data[0][1]['2. Symbol']] * len(ddf.columns), ddf.columns]
errors2 = []

for df in data[1:]:
    ddf2 = df[0].copy()
    ddf2.columns = [[df[1]['2. Symbol']] * len(ddf2.columns), ddf2.columns]
    try:
        ddf = ddf.join(ddf2, how='outer')
    except ValueError:
        errors2.append(df[1]['2. Symbol'])
ddf.head()
```

```python
#
ddf0 = pd.read_csv('data.csv', header=[0,1], index_col=0)
ddf0.head()
```

```python
add_cols = [col for col in ddf.columns if col not in ddf0.columns]
ddf_save = ddf0.join(ddf[add_cols], how='outer')
ddf_save.head()
```
